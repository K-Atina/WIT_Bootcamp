### Title: Web Scraping Project



For web scraping , we will first need to import three libraries: requests (to download the web page itself), BeatifulSoup(for parsing) and Pandas(for organization)

```python
#import required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
page = requests.get(url)

#Parsing
soup = BeautifulSoup(page.text, "html")
```

```python
hockey_table = soup.find("table", class_="table")
# Will look for <table> element with the class "table"
```

```python
#Extracting column Headers
headers = [th.text.strip() for th in hockey_table.find_all("th")]

#Create empty dataframes and store the headers
df = pd.DataFrame(columns=headers)
df.head()
```

```python
# Extracting rows
rows = hockey_table.find_all("tr")

for row in rows[1: ]: #As we know the range usually skips first index
    cols = row.find_all("td")
    data = [col.text.strip() for col in cols]
    if data: #This will make sure we don't have empty rows
      df.loc[len(df)] = data
```

```python
df.head(10)
#To check if there is data
```

```python
#Save data as Hockey.csv
df.to_csv("Hockey.csv", index=False)
```
